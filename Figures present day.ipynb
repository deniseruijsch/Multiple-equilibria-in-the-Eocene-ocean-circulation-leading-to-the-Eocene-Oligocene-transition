```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import xarray as xr
```

```python
ds2 = xr.open_dataset('na pert 1 4deg.overturning.nc')
ds = xr.open_dataset('perturbation 4deg.overturning.nc') 
ds3 = xr.open_dataset('1000_year_4deg.overturning.nc') 
ds4 = xr.open_dataset('1020 year 4deg.overturning 2.nc') 
ds5 = xr.open_dataset('na pert 2 4deg.overturning.nc') 
ds6 = xr.open_dataset('na pert 3 4deg.overturning.nc') 
ds7 = xr.open_dataset('na pert 4 4deg.overturning.nc') 
ds8 = xr.open_dataset('na pert 5 4deg.overturning.nc') 
```

```python
im = ds8['vsf_depth'].isel(Time=-1).plot.pcolormesh(levels=50, add_colorbar=False)
CS = ds8['vsf_depth'].isel(Time=-1).plot.contour(colors='k', linewidths=0.5, levels=10)
plt.clabel(CS, inline=True, fontsize=8)
cb = plt.colorbar(im)
cb.set_label(label='Meridional transport (Sv)')
plt.axis([-80, 80, -4500, -500]) 
plt.xlabel("Meridional coordinate (degrees north)")
plt.ylabel("Vertical coordinate (m)")
plt.title("Contour plot of the MOC at t=1500")
plt.savefig('year 1500 contour', dpi=1020)
```
